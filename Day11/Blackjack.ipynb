## Capstone Project - Blackjack

```python
# import libraries
import art
import random
from IPython.display import clear_output

# function to calculate the score
def calculate_score(card_list):
    return sum(card_list)

# function to deal card
def deal_card(card_list, cards):
    card = random.choice(cards)
    new_list = card_list
    if card == 11: # check if card is Ace
        if (sum(card_list) + card) > 21:
            new_list.append(1)
        else:
            new_list.append(card)
    else:
        new_list.append(card)
    return new_list

# function to print generic end message
def end_message(user_cards, dealer_cards):
    print(f"Your final hand: {user_cards}, final score = {calculate_score(user_cards)}")
    print(f"Dealer's final hand: {dealer_cards}, final score = {calculate_score(dealer_cards)}")
    print()

# function to print generic continue message
def continue_message(user_cards, dealer_cards):
    print(f"Your cards: {user_cards}, current score = {calculate_score(user_cards)}")
    print(f"Dealer's first card: {dealer_cards[0]}")
    print()

# function to compare scores and display winner
def check_winner(user_score, dealer_score):  
    if dealer_score > 21 or user_score > dealer_score:
        print("Congratulations!!!🎉 You win😎")
    elif dealer_score == user_score:
        print("Uh oh, it's a draw!🥴")
    else:
        print("Bummer! You lose😭")
    print()

# blackjack function
def blackjack():
    # ask user to play
    play_game = input("Do you want to play a game of Blackjack?♠️♦️♣️🃏♥️ (yes or no): ").lower()

    # user choose to play
    if play_game == "yes":
        clear_output(wait=False) # clear the screen
        print(art.logo) # print logo when user starts playing
        
        cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10] # card list
        
        # deal first two sets of card
        user_cards = random.choices(cards, k=2)
        dealer_cards = random.choices(cards, k=2)
        
        # calculate the current score
        user_score = calculate_score(user_cards)
        dealer_score = calculate_score(dealer_cards)
        
        if user_score > 21: # bust 
            end_message(user_cards, dealer_cards)
            print("Bust! You lose😭")
        elif dealer_score == 21: # blackjack win
            end_message(user_cards, dealer_cards)
            print("You lose!😭 Dealer wins with a Blackjack.")
        elif user_score == 21: # blackjack win
            end_message(user_cards, dealer_cards)
            print("Blackjack win!!!😎")
        else: # continue playing
            continue_message(user_cards, dealer_cards) # show current cards
            deal_again = input("Type 'yes' to deal another card, or 'no' to end 🎲: ").lower()
    
            # user choose to deal
            while deal_again == "yes":
                # deal a random card
                user_cards = deal_card(user_cards, cards) 
                dealer_cards = deal_card(dealer_cards, cards)
                
                # calculate the current score
                user_score = calculate_score(user_cards)
                dealer_score = calculate_score(dealer_cards)
    
                if user_score > 21: # bust
                    end_message(user_cards, dealer_cards)
                    print("Bust! You lose😭\n")
                    deal_again = "lost"
                else: # continue
                    continue_message(user_cards, dealer_cards)  # show current cards
                    deal_again = input("Type 'yes' to deal another card, or 'no' to end 🎲: ").lower()
    
            # user choose not to deal
            if deal_again == "no":
                while dealer_score <= 16: # dealer continues playing
                    dealer_cards = deal_card(dealer_cards, cards)   
                    dealer_score = calculate_score(dealer_cards)      
                end_message(user_cards, dealer_cards) # show final cards
                check_winner(user_score, dealer_score) # compare scores and display winner

        blackjack() # perform recursion
        
blackjack() # call the blackjack function to begin the game
print(art.logo) # print logo when user stops playing
```
