## Day 4 Project - Rock Paper Scissors

```python
rock = '''
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
'''

paper = '''
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)
'''

scissors = '''
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
'''

#Write your code below this line ðŸ‘‡
import random

game_title = "Rock, Paper, Scissors"

print("*"*len(game_title))
print(game_title)
print("*"*len(game_title))

print(rock + "  " + paper + "  " + scissors)

options = [rock, paper, scissors]
options_title = ["rock", "paper", "scissors"]

user_choice = int(input("Enter your choice (Rock = 0, Paper = 1, Scissors = 2): "))
computer_choice = random.randint(0, 2)

if user_choice == 0 or user_choice == 1 or user_choice == 2:
    print(f"\nYou chose {options_title[user_choice]} {options[user_choice]}\n")
    print(f"Computer chose {options_title[computer_choice]} {options[computer_choice]}\n")
  
    if user_choice == computer_choice:
        print("It is a draw!")
  
    elif (user_choice == 0 and computer_choice == 2) or (user_choice == 2 and computer_choice == 1) or (user_choice == 1 and computer_choice == 0):
        print("You win!!!!")

    else:
        print("Computer wins!!!!")

else:
    print("\nInvalid input! Please enter a valid choice.")
print("*"*50)
```
